```python
y = airbnb["price"]
X = airbnb.loc[:, airbnb.columns!="price"]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

onehot_ftrs = ['neighbourhood_group', 'room_type', 'neighbourhood']
std_ftrs = ['latitude','longitude','minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count','availability_365']

#preprocess the data using onehot encoder and standard scalar encoder for catergorical and continunous data respectively.
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant',fill_value='missing')),
    ('onehot', OneHotEncoder(sparse=False,handle_unknown='ignore'))]) 

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('std', StandardScaler())])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, std_ftrs),
        ('cat', categorical_transformer, onehot_ftrs)])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder, MinMaxScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
import math
from sklearn.metrics import make_scorer, r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV


def MLpipe_KFold(X, y, preprocessor, ML_algo, param_grid):
#This def takes preprocessor,dataset, ML algorithm and parameters as input. Then uses GridSearchCV to select best parameters
#r2_score is used to evaluate the test data set. The grid with best parameters are returned along with the r2 scores.
    grid_lst = []
    bst_scores = []
    test_scores = []
    random_state = [2,4,10,15,28,35]#looping through 5 random states
    for i in random_state:
        X_other, X_test, y_other, y_test = train_test_split(X, y, test_size = 0.2, random_state=i)
        kf = KFold(n_splits=5)#split into 5 folds
        pipe = Pipeline(steps=[('preprocessor', preprocessor), ('ML_algo', ML_algo)])
        grid = GridSearchCV(pipe, param_grid=param_grid, scoring = make_scorer(r2_score) ,cv=kf, return_train_score = True, n_jobs=-1, verbose=True)
        grid.fit(X_other, y_other)
        y_pred = grid.predict(X_test)
        test_scores.append(r2_score(y_test, y_pred))
        grid_lst.append(grid.best_params_)
    return grid_lst, test_scores
```

```python
from sklearn.linear_model import Lasso
import warnings
warnings.filterwarnings("ignore")

#Lasso with parameters is used to train the model. Model with best parameters is selected. The mean of r2 score is calculated.
param_grid = {'ML_algo__alpha':[100.0, 35.938136638046274, 12.91549665014884, 4.641588833612779, 1.668100537200059, 0.5994842503189408, 0.21544346900318845, 0.07742636826811278, 0.027825594022071257, 0.01], 'ML_algo__max_iter':[100000000], 'ML_algo__random_state':[10]} 
model1, test_scores1 = MLpipe_KFold(X, y, preprocessor, Lasso(), param_grid)
print("-----------------------------")
for m in model1:
    print(m)
print("-----------------------------")
for s in test_scores1:
    print(s)
print("-----------------------------")
print('r2_score:',"mean",np.around(np.mean(test_scores1),3),'+/-',np.around(np.std(test_scores1),3),"standard deviation")
```

```python
import xgboost
from sklearn.model_selection import ParameterGrid
import warnings
warnings.filterwarnings("ignore")


random_state = [5, 10, 13, 20, 25]
all_val_scores = []
all_train_scores = []
all_test_scores = []

train_scores_mean = []
val_scores_mean = []
test_scores_mean = []

train_scores_std = []
val_scores_std = []
test_scores_std = []

for i in random_state:#looping through 5 random states
    X_other1, X_test1, y_other1, y_test1 = train_test_split(X, y, test_size = 0.2, random_state=i)    
    kf = KFold(n_splits=5)#split into 5 folds
    for train_index, val_index in kf.split(X_other1,y_other1):
        X_train1 = X_other1.iloc[train_index]
        y_train1 = y_other1.iloc[train_index]
        X_CV = X_other1.iloc[val_index]
        y_CV = y_other1.iloc[val_index]
            
        
    pipeline_fit = Pipeline(steps=[('preprocessor', preprocessor)]) 
        
    X_train1 = pipeline_fit.fit_transform(X_train1)
                        
    X_CV = pipeline_fit.transform(X_CV)

    X_test1 = pipeline_fit.transform(X_test1)
        
    depth_add = []
    sm_add = []
    test_scores = []
    train_scores = []
    val_scores = []
    #loop trough max_depth and subsample parameters and find the best set of parameters.
    max_dep = [1, 2, 3, 6]
    subsample = [0.4, 0.5, 0.66]
    for md in max_dep: 
        for sm in subsample:
            param_grid = {"learning_rate": [0.03],
                              "n_estimators": [10000],
                              "max_depth": [md],
                              "colsample_bytree": [0.9],              
                              "subsample": [sm]}

            XGB = xgboost.XGBRegressor()
            XGB.set_params(**ParameterGrid(param_grid)[0])
            XGB.fit(X_train1,y_train1,early_stopping_rounds = 40,eval_set=[(X_CV, y_CV)], verbose=False)
            y_CV_pred = XGB.predict(X_CV)
            ms1 = r2_score(y_CV,y_CV_pred)#evaluate validation scores
            val_scores.append(ms1)#append validation scores
           
            y_train_pred = XGB.predict(X_train1)
            ms2 = r2_score(y_train1,y_train_pred)#predict train scores
            train_scores.append(ms2)#append tain scores
           
            
            y_test_pred = XGB.predict(X_test1)
            ms3 = r2_score(y_test1, y_test_pred)#predict test scores
            test_scores.append(ms3)#append test scores
            depth_add.append(md)#append depth
            sm_add.append(sm)#append subsample
                                    
            
    #mean of the scores        
    train_scores_mean.append(np.mean(train_scores))
    val_scores_mean.append(np.mean(val_scores))
    test_scores_mean.append(np.mean(test_scores))
    #standard deviation of the scores        
    train_scores_std.append(np.std(train_scores))
    val_scores_std.append(np.std(val_scores))
    test_scores_std.append(np.std(test_scores))
    #store all the scores
    all_val_scores.append(val_scores_mean)
    all_train_scores.append(train_scores_mean)
    all_test_scores.append(test_scores_mean)
            
    print("********************************************")
    #Display the model with best parameters for each random state.
    print("RANDOM STATE: ", i)
    print("Best test scores: ", np.max(test_scores), " ---->   Best max_depth: ", depth_add[np.array(test_scores).argmax()], " Best subsample: ", sm_add[np.array(test_scores).argmax()], "learning_rate: 0.03,  n_estimators: 10000, colsample_bytree: 0.9")
    #write the results into the file
print(" score:(mean)",np.mean(all_test_scores),'+/-',np.std(all_test_scores),"(standard deviation)")
```
